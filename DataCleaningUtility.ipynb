#### Data Cleaning Utility

```python
#Step 1: Import Required Library
import pandas as pd
```

```python
#Step 2: Read the CSV File
df = pd.read_csv("Data.CSV")
print(df)
```

```python
#Step 3: Understand the Dataset

df.info()

#Shows column names,data types
#Shows missing values
```

```python
#Step 4: Check Missing Values

df.isnull().sum()

#Tells how many missing values are present in each column.
```

```python
 df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
#step 4:Covert age to numeric
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
#Fill missing age
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
#step 5:Convert salary to Numeric
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
```

```python
#Fill missing salary
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
#step 6: Handle Remaining Missing Dates

# Clean column names
df.columns = df.columns.str.strip()

# Fix Join Date
df['Join Date'] = pd.to_datetime(df['Join Date'], errors='coerce')
df['Join Date'] = df['Join Date'].ffill()

#Uses previous valid date to fill missing date
```

```python
#Step 7: Remove Duplicate Records
df.drop_duplicates(inplace=True)

#Removes repeated rows
#Ensures data accuracy
```

```python
#Step 8: Standardize Column Names

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
#Step 9: Final Cleaned Data
print(df)
```

```python
# Step 10: Save Cleaned Dataset

df.to_csv("Clean.csv", index=False)
```
